```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine data 
pyber_df = pd.merge(ride_data, city_data, on="city", how="left")
pyber_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the data frames for the city type(urban_rural,suburban) and then group the data frames by city
urban_city_type = pyber_df[pyber_df["type"] == "Urban"].groupby([pyber_df["city"]])
rural_city_type = pyber_df[pyber_df["type"] == "Rural"].groupby([pyber_df["city"]])
suburban_city_type = pyber_df[pyber_df["type"] == "Suburban"].groupby([pyber_df["city"]])

# Urban calculations of ride count x, fare average y, driver count average Z
x_urban = urban_city_type["ride_id"].count()
y_urban = urban_city_type["fare"].mean()
z_urban = urban_city_type["driver_count"].mean()

# Rural calculations of ride count x, fare average y, driver count average Z
x_rural = rural_city_type["ride_id"].count()
y_rural = rural_city_type["fare"].mean()
z_rural = rural_city_type["driver_count"].mean()

# Suburban calculations of ride count x, fare average y, driver count average z
x_suburban = suburban_city_type["ride_id"].count()
y_suburban = suburban_city_type["fare"].mean()
z_suburban = suburban_city_type["driver_count"].mean()

# Build the scatter plots for each city types
plt.scatter(x_urban, y_urban, label = "Urban", s=z_urban * 10, color=["gold"], edgecolor="black", alpha = 0.75, marker="o")
plt.scatter(x_rural, y_rural, label = "Rural", s=z_rural * 10, color=["skyblue"], edgecolor="black", alpha = 0.75, marker="o")
plt.scatter(x_suburban, y_suburban, label = "Suburban", s=z_suburban  * 10, color=["coral"], edgecolor="black", alpha = 0.75, marker="o")
plt.grid()

# Create a legend
legend = plt.legend(fontsize = 8, title= "City Type", loc="best") 

# Incorporate the other graph properties
plt.xlabel("Total Number of Rides")
plt.ylabel("Average Cost of Fare")
plt.title("Pyber Ride Sharing Data")

# Save image
plt.savefig("PyberRideSharingData.png", bbox_inches="tight")

```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
type_groupped = pyber_df.groupby(['type'])
fare_sum = type_groupped['fare'].sum()

# Build Pie Chart
labels = ["Rural","Suburban","Urban"]
explode = (0, 0, 0.1)
colors = ["skyblue", "coral", "gold"]
plt.pie(fare_sum, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Fares")

# Save Figure
plt.savefig("TotalFares.png")
```

```python
# Show Figure
plt.show()
```

```python
# Calculate Type Percents
rides_count = type_groupped['ride_id'].count()

# Build Pie Chart
labels = ["Rural","Suburban","Urban"]
explode = (0, 0, 0.15)
colors = ["skyblue", "coral", "gold"]
plt.pie(rides_count, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Rides by City Type")

# Save Figure
plt.savefig("TotalRides.png")
```

```python
# Show Figure
plt.show()
```
